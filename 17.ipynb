```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

# c nh
image_path = 'C:/Users/Administrator/Documents/Hieu2024/ThiGiacMayTinh/BaiTapTGMT/img/anime.jpg'
image = Image.open(image_path)

# nh xm
gr_image = image.convert('L')

# Chuyn nh xm v mng
image_a = np.array(gr_image)

# Tnh gradient theo hng x
gradient_x = ndimage.sobel(image_a, axis=1, mode='constant')

# Hin th nh
fig, axs = plt.subplots(2,1, figsize=(10, 8))

plt.subplot(2,1,1)
plt.imshow(gr_image, cmap='gray')
plt.title('nh xm gc')

plt.subplot(2,1,2)
plt.imshow(gradient_x, cmap='gray')
plt.title('nh xm gc')

plt.show()
```
