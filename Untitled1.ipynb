```python
# Import lớp Image
from PIL import Image
```

```python
# Đường dẫn đến hình ảnh
image_path = dir +'/1.PNG'
print(image_path)
```

```python
# Sử dụng Image.open() đọc ảnh
picture = Image.open(image_path)
```

```python
# Hiển thị hình ảnh
display(picture)
```

```python
# In một số thông tin
# In định dạng ảnh
print("Định dang ảnh: ", picture.format)
```

```python
# In ra size
print("Kích thước hình ảnh: ", picture.size)
```
