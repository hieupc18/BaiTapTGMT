```python

# Kích thước ban đầu:
print(imgs.size)
```

```python
# Thay đổi kích thước
new_size = (1000,1000)
new_image = imgs.resize(new_size)
display(new_image)
# Task: lưu ảnh
```

```python
# Thay đổi kích thước
thumbnail_size = (400,400)
imgs.thumbnail(thumbnail_size)
display(imgs)
# Task: lưu ảnh
```

```python

# Chuyển từ ảnh RGB sang ảnh xám
gray_image = imgs.convert("L")
display(gray_image)
```

```python
# Phân RGB thành 3 tầng màu khác nhau
red_band, green_band, blue_band = imgs.split()
display(red_band)
display(green_band)
display(blue_band)
```

```python
# Trộn ảnh
merged_image =  Image.merge("RGB", (red_band, green_band, blue_band))
display(merged_image)
```
